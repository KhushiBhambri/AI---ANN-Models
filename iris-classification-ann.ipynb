```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df = pd.read_csv('/kaggle/input/iris-flower-dataset/IRIS.csv')
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore')
encoded = ohe.fit_transform(df[['species']]).toarray()
```

```python
encoded = pd.DataFrame(encoded)
encoded.head()
```

```python
df = df.join(encoded)
df.drop('species',axis=1,inplace=True)
df.head()
```

```python
df[df.isnull()].count()
```

```python
X = df.iloc[:,:4]
X.head()
```

```python
y = df.iloc[:,4:]
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.7,random_state=101)
```

```python
ann = tf.keras.models.Sequential()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
ann.add(tf.keras.layers.Dense(4,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(4,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(4,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(4,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(3,activation='softmax'))
```

```python
ann.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100,batch_size=7)
```

```python
y_pred = ann.predict(X_test)
```
